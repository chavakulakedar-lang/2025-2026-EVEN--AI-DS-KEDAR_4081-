```python

# Selection data
import pandas as pd
data = {
    "ID": [101,102,103,104,105],
    "Name": ["Alice","Bob","Charlie","David","Eva"],
    "Age": [22,25,23,24,26],
    "Marks": [85,78,92,88,90]
}
df = pd.DataFrame(data)
print(df)
print(df["Name"])
print(df["Marks"])
print("location 0", df.iloc[0])
print("location 1 to 4", df.iloc[1:4])
print("location 1 to 3", df.iloc[0:3,1:3])
```

```python
# Selection Sort
n = int(input("enter number of elements:"))

arr = []
for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

print("Original array:", arr)

for i in range(n-1):
    min_index = i
    for j in range(i+1,n):
        if arr[j] < arr[min_index]:
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print("Sorted array:", arr)
```

```python
# Bubble Sort
n = int(input("enter number of elements:"))

arr = []
for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

print("Original array:", arr)

for i in range(n-1):
    for j in range(n-1-i):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print("Sorted array:", arr)
```

```python
# Linear Search
n = int(input("enter number of elements:"))

arr = []
for i in range(n):
    arr.append(int(input(f"enter element {i}:")))

key = int(input("enter element to search:"))

found = False
for i in range(n):
    if arr[i] == key:
        print(f"element {key} found at position {i+1}")
        found = True
        break

if not found:
    print("element not found")
```

```python
# Binary Search
n = int(input("enter number of elements:"))

arr = []
for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

arr.sort()
print("sorted list:", arr)

key = int(input("enter element to search:"))

low = 0
high = n-1
found = False

while low <= high:
    mid = (low + high)//2

    if arr[mid] == key:
        print(f"element {key} found at position {mid+1}")
        found = True
        break
    elif key < arr[mid]:
        high = mid-1
    else:
        low = mid+1

if not found:
    print("element not found")
```
