```python

numbers = {1, 2, 3, 4, 4, 5}
print(numbers)  # Duplicates removed automatically

# Adding elements
numbers.add(6)
print(numbers)

# Removing elements
numbers.remove(3)
print(numbers)

# Checking membership
print(2 in numbers)   # True

# Looping through a set
for num in numbers:
    print(num)

# Set operations
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.union(set2))        # {1,2,3,4,5}
print(set1.intersection(set2)) # {3}
print(set1.difference(set2))   # {1,2}
```
